# RAPIDAPI to Get Stock Prices
* This call will let you get stock price of any ticker that you have a symbol for

```python
import pandas as pd
import matplotlib as plt
```

```python
import requests

url = "https://apidojo-yahoo-finance-v1.p.rapidapi.com/market/get-charts"

querystring = {"symbol":"CEO","interval":"1mo","range":"5y","region":"US"}

headers = {
    'x-rapidapi-key': "a03ce9303cmshe8ef211d48354c9p1ae5a5jsn7f23105c1c5d",
    'x-rapidapi-host': "apidojo-yahoo-finance-v1.p.rapidapi.com"
    }

response = requests.request("GET", url, headers=headers, params=querystring)

print(response.text)
```

```python
import json
json_data = response.text
```

```python
json_object = json.loads(json_data)
json_formatted_str = json.dumps(json_object, indent=2)
print(json_formatted_str)
```
