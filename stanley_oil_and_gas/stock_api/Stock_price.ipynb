```python
import yfinance as yf

msft = yf.Ticker("CEO")

# get stock info
print(msft.info)

# get historical market data
hist = msft.history(period="5y")
```

```python
hist
```

```python
import matplotlib.pyplot as plt
import seaborn
```

```python
ax = hist['Close'].plot(figsize=(16,9))
ax.set_xlabel("Date")
ax.set_ylabel("Price")
ax.set_title("CEO 5 Year Stock Price")
```

```python
fig = ax.get_figure()
fig.savefig('output.png')
```
